```python
import numpy as np

def similar(x, y, df2):
    mostSimilar = x, y, 1000
    diff = 10, 10
    
    for i in range(len(df2)):
        if (abs(x-df2.X.values[i]) < diff[0]) & (abs(y-df2.Y.values[i]) < diff[1]):
            mostSimilar = df2.X.values[i], df2.Y.values[i], df2.time.values[i]
            diff = abs(x-df2.X.values[i]), abs(y-df2.Y.values[i])
                                            
    return mostSimilar
    
#print similar(481,474, data)

def velocity(df1, df2):
    velFrame = []
    for i in range(len(df1)):
        x = df1.X.values[i]
        y = df1.Y.values[i]
        sim = similar(x, y, df2)
        #print sim
        vel = 2000000000*(np.sqrt( (sim[0] - x)**2 + (sim[1] - y)**2))/abs(int(df1.time.values[0] - sim[2]))
        velFrame.append([vel, x, y, sim[0], sim[1]])
    return velFrame
    #print " "
```
